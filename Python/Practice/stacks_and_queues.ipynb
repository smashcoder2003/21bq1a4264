```python
# To create a stack
class Stack:
    # Creating a stack object initializes a list
    def __init__(self):
        self.stack = []
        self.top = -1
     # Checks whether the stack is empty.
    def is_empty(self):
        return self.top == -1

    # Pushes and element on to the stack
    def push(self,data):
        self.top += 1
        self.stack.append(data)
        return f"The element {data} has been pushed on to the stack."

    # Pops and element from the stack
    def pop(self):
        if self.is_empty():
            raise StackUnderFlowError("The Stack is already empty!!")
        else:
            self.top -= 1
            return f"The element {self.stack.pop()} has been popped from the stack."

    # Returns the top of the stack
    def peek(self):
        return self.top

    # Prints all the elements in the stack
    def print_stack(self):
        if self.is_empty():
            print("Stack is empty !!")
        for i in range(len(self.stack)):
            if i == len(self.stack) - 1:
                print(self.stack[i])
            else:
                print(f"{self.stack[i]} -->", end=" ")

class StackUnderFlowError(Exception):
    pass

if __name__ == '__main__':
    stack = Stack()
    print(stack.peek())
    print(stack.push(10))
    print(stack.push(15))
    print(stack.push(64))
    print(stack.push(26))
    stack.print_stack()
    print(stack.pop())
    print(stack.pop())
    stack.print_stack()
    print(stack.pop())
    stack.print_stack()
    stack.pop()
    stack.pop()
```

```python
# To create a Queue
class Queue:
    def __init__(self):
        self.Queue = []
        self.top = -1

    # Adds items / elements to the queue
    def enqueue(self,data):
        self.Queue.append(data)
        self.top += 1
        return  f"The element {data} has been added to the Queue."

    # Checks whether the queue is empty
    def is_empty(self):
        return self.top == -1

    # Removes an item from the queue
    # Throws and exception if the user tries to dequeue when the Qeueue is already empty
    def dequeue(self):
        if self.is_empty():
            raise QueueUnderFlowError("The Queue is already empty !!")
        else:
            self.top -= 1
            return f"The element {self.Queue.pop(0)} has been dequeued from the Queue."

    # Returns the top of the stack
    def peek(self):
        return self.top

    # To print the items of a stack
    def print_stack(self):
        if self.is_empty():
            print("Stack is empty !!")
        for i in range(len(self.stack)):
            if i == len(self.stack) - 1:
                print(self.stack[i])
            else:
                print(f"{self.stack[i]} -->", end=" ")


class QueueUnderFlowError(Exception):
    pass


if __name__ == '__main__':
    queue = Queue()
    print(queue.peek())
    print(queue.enqueue(25))
    queue.enqueue(34)
    print(queue.dequeue())
    print(queue.dequeue())
    queue.dequeue()
```
